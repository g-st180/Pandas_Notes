```python
def ismovie(type):
    if type=='Movie':
        return 1
    else:
        return 0
    
df['Movie']=df['Type'].apply(ismovie)
```

```python
def isTvShow(type):
    if type=='TV Show':
        return 1
    else:
        return 0
    
df['Tv_Show']=df['Type'].apply(isTvShow)
```

```python
df['Type']=df['Type'].map({'Movie':1,'TV Show':-1})
```
